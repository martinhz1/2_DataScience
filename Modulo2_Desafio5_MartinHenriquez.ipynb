```python
#creamos las columnas
df[['Salario Estimado Mínimo', 'Salario Estimado Máximo']] = df['Salary Estimate'].apply(lambda x: pd.Series(x.split('-')))
```

```python
df
```

```python
# Convertir las columnas a tipo numérico si es necesario
df['Salario Estimado Mínimo'] = pd.to_numeric(df['Salario Estimado Mínimo'])
df['Salario Estimado Máximo'] = pd.to_numeric(df['Salario Estimado Máximo'])
```

### Etiquetamos los valores de la columna Size por tipo de empresa en base a su numero de empleados.

```python
size_mapping = {
    '10000+ employees': 'MegaEmpresas',
    '5001 to 10000 employees': 'Grandes Empresas',
    '1001 to 5000 employees': 'Medianas Empresas',
    '201 to 500 employees': 'Pequeñas Empresas',
    '51 to 200 employees': 'Pequeñas Grandes Empresas',
    '501 to 1000 employees': 'MicroEmpresas',
    'Unknown': 'Empresa sin información'
}
```

```python
df['Size'] = df['Size'].replace(size_mapping)
```

```python
df
```

```python
pivot_table2 = pd.pivot_table(df, 
                             index='Size', 
                             values=['Salario Estimado Máximo', 'Salario Estimado Mínimo'], 
                             aggfunc='mean')
```

```python
print("Tabla pivote con la media del salario estimado máximo y mínimo por tamaño de empresa:")
print(pivot_table2)
```
