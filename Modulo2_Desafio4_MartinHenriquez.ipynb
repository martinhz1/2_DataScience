### 1. Crear Dataframe desde archivo .csv

```python
import pandas as pd
import numpy as np 

df = pd.read_csv('/Users/martinhenriquez/Downloads/resumen_resultados_astronautas.csv')
```

```python
df.info()

#Contamos con 32 registros y 10 columnas
```

### El IMC es una medida importante que nos ayuda a determinar si una persona tiene un peso saludable para poder ser un astronauta. Se calcula siguiendo la formula peso/(estatura^2), donde peso está en kg y estatura en metros. 

```python
# Creo una nueva columna con la altura en metros para realizar calculo del IMC 

df["altura_m"] = df["altura"] / 100
```

```python
#Creo columna IMC con la nueva columna altura_m

df["IMC"] = df["peso"] / (df["altura_m"] ** 2)
df
```

```python
#Agregamos columna promedio_evaluaciones

df['promedio_evaluaciones'] = (df['evaluacion_fisica'] + df['evaluacion_psicologica'] + df['evaluacion_adaptativa'] + df['evaluacion_medica']) / 4
```

```python
#Nuevo dataframe donde solo iran aquellos que estén dentro del rango de IMC requerido

df2 = df.loc[(df['IMC'] >= 18.5) & (df['IMC'] <= 24.9)]
```

```python
df2

#Vemos que en el nuevo df hay vacios en el index (7,19,22,26) que corresponden a aquellos con un IMC fuera del rango
```

```python
#Reset index del nuevo dataframe
df2 = df2.reset_index(drop=True)
df2
```

```python
# Nuevo dataframe en base al df filtrado segun IMC donde solo iran aquellos con el promedio de evaluaciones requerido
df3 = df2.loc[(df2['promedio_evaluaciones'] >= 87)]
```

```python
df3

#Vemos que en el nuevo df hay vacios en el index que corresponden a aquellos casos que no cumplen el promedio de evaluaciones
```

```python
#reset index del dataframe final 
df3 = df3.reset_index(drop=True)
```

```python
df3
```

```python
#Reemplazamos los valores "No" de la columna 'califica' dentro del mismo Dataframe

df3['califica'].replace("No", "Si", inplace=True)
```
